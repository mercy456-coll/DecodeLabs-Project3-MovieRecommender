```python
# Movie dataset - each movie has a title and descriptive tags
# Tags represent genre, mood and themes used for recommendation matching

movies = [
    # Romance & Emotional
    {"title": "The Notebook", "tags": "romance emotional heartbreak classic drama"},
    {"title": "Five Feet Apart", "tags": "romance emotional illness heartbreak drama"},
    {"title": "A Walk to Remember", "tags": "romance emotional inspiration classic drama"},
    {"title": "Titanic", "tags": "romance tragedy emotional classic drama"},
    {"title": "Crazy Rich Asians", "tags": "romance comedy culture wealth fun"},

    # Thriller & Mystery
    {"title": "Glass Onion A Knives Out Mystery", "tags": "mystery crime comedy twist suspense"},
    {"title": "Knives Out", "tags": "mystery crime comedy suspense"},
    {"title": "The Housemaid", "tags": "psychological thriller twist gothic drama"},
    {"title": "Smile", "tags": "horror psychological dark disturbing thriller"},
    {"title": "Get Out", "tags": "horror psychological race thriller suspense"},

    # Action & Adventure
    {"title": "Black Panther", "tags": "action superhero culture sci-fi adventure"},
    {"title": "Avengers Endgame", "tags": "action superhero sci-fi emotional adventure"},
    {"title": "RRR", "tags": "action epic friendship drama India"},
    {"title": "Spider Man No Way Home", "tags": "action superhero sci-fi fun adventure"},
    {"title": "Top Gun Maverick", "tags": "action military inspiration drama emotional"},
    {"title": "Dune", "tags": "sci-fi epic fantasy adventure drama"},

    # Comedy & Fun
    {"title": "Roommates", "tags": "comedy friendship college identity fun"},
    {"title": "What Men Want", "tags": "comedy romance supernatural fun drama"},
    {"title": "Barbie", "tags": "animation comedy fantasy empowerment identity"},

    # Drama & Inspiration
    {"title": "Parasite", "tags": "thriller crime social drama dark"},
    {"title": "The Woman King", "tags": "action history inspiration empowerment drama"},
    {"title": "Whiplash", "tags": "drama music ambition inspiration psychological"},

    # Animation & Fantasy
    {"title": "KPop Demon Hunters", "tags": "animation fantasy music Korean empowerment friendship"},

    # International
    {"title": "Squid Game The Movie", "tags": "thriller survival social Korean dark"},
    {"title": "Money Heist The Movie", "tags": "crime heist suspense Spanish action"},
]
```

```python
# Import necessary libraries

# pandas - to organize movies into a dataframe 
import pandas as pd

# TfidfVectorizer - converts text tags into weighted number vectors
from sklearn.feature_extraction.text import TfidfVectorizer

# cosine_similarity - measures how similar two vectors are
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Convert movies list into a DataFrame
df = pd.DataFrame(movies)
print(df.head(3))
```

```python
# Convert movie tags into TF-IDF vectors
# This transforms text into weighted numbers the machine can understand
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['tags'])

print("TF-IDF Matrix Shape:",tfidf_matrix.shape)
```

```python
# Collect minimum 3 interests from the user
print("Welcome to KorexBot Movie Recommender!")
print()
print("=" * 40)
print()
print("Enter your movie interests separated by spaces.")
print("Minimum 3 interests required.")
print("Example: romance comedy thriller fun action")
print()

user_input = input("Your interests: ").lower().strip()
interests = user_input.split()

# Validate minimum 3 interests 
if len(interests) < 3:
    print("Please enter at least 3 interests!")
else:
    user_profile = user_input
    print()
    print(f"You entered {len(interests)} interests: {user_profile}")
```

```python
# Transform user profile using the SAME tfidf vocabulary
user_vector = tfidf.transform([user_profile])

# Calculate cosine similarity between user profile and all movies
similarity_scores = cosine_similarity(user_vector, tfidf_matrix)

# Convert scores to a readable list
scores = list(enumerate(similarity_scores[0]))
```

```python
# Sort movies by similarity score in descending order
sorted_scores = sorted(scores, key = lambda x: x[1], reverse = True)

# Display Top 5 recommendations with matched interests
print("Top 5 Movie Recommendations For You:")
print("=" * 40)
print()

for i, (movie_index, score) in enumerate(sorted_scores[:5]):
    movie_title = df['title'][movie_index]
    movie_tags = df['tags'][movie_index]

    # Find which of your interests matched this movie
    matched = [word for word in interests if word in movie_tags]
    
    print(f"{i+1}. {movie_title}")
    print(f"   Match Score: {round(score * 100, 2)}%")
    print(f"   Matched Interests: {', '.join(matched) if matched else 'General match'}")
    print()
```

## Recommendation System Insight

### How KorexBot Movie Recommender Works

This recommendation system is built on Content-Based Filtering —
it matches what YOU like directly to what movies contain,
without needing any other user's data or behavior.

### The Pipeline

1. **Ingestion** — The user enters a minimum of 3 interests
   describing the kind of movies they enjoy

2. **Vectorization** — Both the user profile and movie tags
   are converted into TF-IDF vectors. This means specific
   and unique words carry more weight than common ones

3. **Scoring** — Cosine Similarity measures the angle between
   the user vector and every movie vector. The smaller the
   angle, the higher the match score

4. **Sorting & Filtering** — All 25 movies are ranked by
   their similarity score and only the Top 5 are displayed
   to prevent choice overload

### Why Content-Based Over Collaborative Filtering?

Content-Based Filtering was chosen because it works
immediately — it does not need historical user data or
other users' behavior. It maps preferences directly
to item attributes with absolute precision.

### What the Match Score Means

- **50%+ match** → Strong alignment with your interests
- **30-50% match** → Partial alignment — some interests matched
- **Below 30%** → General match — worth exploring but not a
  perfect fit

### Key Takeaway

The system proves that intelligent recommendations do not
require complex neural networks. Pure mathematical similarity
— TF-IDF weighting combined with Cosine Similarity — is
enough to build a recommendation engine that feels personal
and precise.
